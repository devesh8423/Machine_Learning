```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
#Loading data
df=pd.read_csv("50_Startups.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
df.describe().round(2).T
```

```python
sns.pairplot(df);
```

# Creating a correlation matrix

```python
df.corr(numeric_only=True)
```

# Extracting feature Array and Target Column

```python
x=df[["R&D Spend","Administration","Marketing Spend"]] # Feature Array
```

```python
y=df["Profit"] # Target Column
```

# Spliting Data Into Train And Test

```python
x_train,x_test,y_train,y_test=(train_test_split(x,y,test_size=0.20,random_state=42))
```

```python
print("Total",len(x))
print("x_train",len(x_train))
print("x_test",len(x_test))
```

# Building the Model

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

# Making Prediction

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred=lr.predict(x_test)
y_pred[:5]
```

# Measuring Performance of the model

```python
print("Accuracy_score_test",lr.score(x_test,y_test))
```

```python
print("Accuracy_score_train",lr.score(x_train,y_train))
```

# Checking Errors

```python
print("mean_absolue_error",mean_absolute_error(y_test,y_pred))
print("mean_squared_error",mean_squared_error(y_test,y_pred))
```
